```python
import os
import wfdb
import numpy as np
import pandas as pd

from tensorflow.keras.models import load_model
```

```python
npy_dir = "../data/npy_data_mitbih"
model_dir = "../models"

X_train = np.load(os.path.join(npy_dir, "X_train_mitbih.npy"))
X_test = np.load(os.path.join(npy_dir, "X_test_mitbih.npy"))
y_train = np.load(os.path.join(npy_dir, "y_train_mitbih.npy"))
y_test = np.load(os.path.join(npy_dir, "y_test_mitbih.npy"))

autoencoder_model = load_model(os.path.join(model_dir, "autoencoder_resampled.keras"), compile = False)
beat_identification_model = load_model(os.path.join(model_dir, "labeling_v2_spatialdropout.keras"), compile = False)
```

```python
def normalize(X):
    mean = X.mean(axis=1, keepdims=True)
    std  = X.std(axis=1,  keepdims=True) + 1e-8
    return (X - mean) / std, mean, std  # devolvemos también mean y std

X_train_norm, mean_tr, std_tr = normalize(X_train)
X_test_norm, mean_tr, std_tr = normalize(X_test)

print(f"X_train es: {type(X_train)}")
print(f"X_train_norm es: {type(X_train_norm)}")
```

```python
X_train_denoised = autoencoder_model.predict(X_train_norm)
X_test_denoised = autoencoder_model.predict(X_test_norm)
```
